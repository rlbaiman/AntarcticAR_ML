```python
# Import necessary libraries
import numpy as np
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.feature import NaturalEarthFeature
import matplotlib.path as mpath
```

# Step 1:
## Calculate percentiles of T2M and PRECSN using Make_Y_data_1.py and Make_Y_data_1.script

### Note: this is done in parallel by splitting up into months. Memory limitations do not allow doing this as one file

# Step 2:

## Create hemisphere dataset (in yearly files) for T2M and PRECSN using Make_Y_data2.py and Make_T_data_2.script

### Each year file timesteps that feature 98th percentile snow rate or 2m temperature. Top 2mt is labeled as 1, top precsn is labeled as 10

# Step 3:

# Make one results file per year using codes: 100 for AR + 10 for top snowfall rate + 1 for top 2mT

```python
for year in range(2010,2020):
    #AR mask
    ar = xr.open_mfdataset('/projects/reba1583/Research2/data/AR_catalogue/'+str(year)+'*').sel(lat = slice(-90,0))
    ar = ar.resample(time="6H").max()
    ar = ar.interp(lon = np.arange(-180,181,1), lat = np.arange(-90,1,1))
    ar = xr.where(ar.ar_binary_tag>0,100,0)
    ar = ar.sel(time = ar.where(ar>0, drop = True).time)
    ar = ar.rename('Y')
    # 98th percentile T2M
    t = xr.open_mfdataset('/rc_scratch/reba1583/Y_variables_98th/T2M'+str(year)+'*')
    t = t.__xarray_dataarray_variable__.rename('Y')
    # 98th percentile PRECSN
    p = xr.open_mfdataset('/rc_scratch/reba1583/Y_variables_98th/PRECSN'+str(year)+'*')
    p = p.__xarray_dataarray_variable__.rename('Y')

    year_times = xr.open_mfdataset('/pl/active/ATOC_SynopticMet/data/ar_data/Research2/3hrly_merra2_hemisphere/'+str(year)+'*')
    year_times = year_times.time.resample(time="6H").min()
    
    Y_dataset = xr.DataArray(np.zeros((len(year_times),len(t.lat), len(t.lon))),
                       dims = ['time','lat','lon'],
                       coords = dict(time = year_times,
                                    lat = t.lat,
                                    lon = t.lon),
                         name = 'Y')
    for time in range(len(Y_dataset.time)):
        if Y_dataset.time[time].isin(t.time):
            Y_dataset[time,:,:] = t.sel(time = Y_dataset.time[time]).values
        if Y_dataset.time[time].isin(p.time):
            Y_dataset[time,:,:] = Y_dataset[time,:,:] + p.sel(time = Y_dataset.time[time]).values
        if Y_dataset.time[time].isin(ar.time):
            Y_dataset[time,:,:] = Y_dataset[time,:,:] + ar.sel(time = Y_dataset.time[time]).values

    Y_dataset.to_netcdf('/rc_scratch/reba1583/yrly_Y_files/'+str(year))
    print(year)
```
